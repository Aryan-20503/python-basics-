### **Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.**

(Hint: Compare their scope, techniques, and applications for each.)



 **Difference Between AI, ML, DL, and Data Science**

Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science are closely related fields, but each has a different **scope, set of techniques, and applications**.

---

 **1. Artificial Intelligence (AI)**

* **Scope:** AI is the broadest concept that refers to the simulation of human intelligence in machines. It aims to create systems that can think, reason, learn, and make decisions like humans.
* **Techniques:** AI uses techniques such as rule-based systems, knowledge representation, search algorithms, expert systems, and natural language processing (NLP).
* **Applications:** AI is applied in robotics, speech recognition (like Siri and Alexa), autonomous vehicles, game playing (e.g., Chess engines), and recommendation systems.

---

 **2. Machine Learning (ML)**

* **Scope:** ML is a **subset of AI** that focuses on developing algorithms that enable computers to **learn from data** and improve their performance without being explicitly programmed.
* **Techniques:** Common techniques include supervised learning (regression, classification), unsupervised learning (clustering, dimensionality reduction), and reinforcement learning.
* **Applications:** ML is used in spam detection, fraud detection, stock market prediction, personalized recommendations, and sentiment analysis.

---

 **3. Deep Learning (DL)**

* **Scope:** DL is a **subset of ML** that uses **artificial neural networks** with multiple layers to automatically learn complex patterns from large amounts of data.
* **Techniques:** It relies on neural networks like Convolutional Neural Networks (CNNs) for image processing and Recurrent Neural Networks (RNNs) for sequential data such as speech or text.
* **Applications:** DL powers facial recognition, autonomous driving, language translation, and medical image analysis.

---

 **4. Data Science**

* **Scope:** Data Science is a **broader interdisciplinary field** that combines statistics, data analysis, and machine learning to extract meaningful insights from structured and unstructured data.
* **Techniques:** It uses tools like Python, R, SQL, data visualization, statistical analysis, and ML algorithms.
* **Applications:** Data Science is applied in business analytics, healthcare analytics, marketing optimization, and social media trend analysis.

---

### **Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?**
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.





 **1. Overfitting**

* **Definition:** Overfitting occurs when a model learns the **training data too well**, capturing noise, random fluctuations, and irrelevant patterns instead of the actual underlying trend.
* **Effect:** Such a model performs very well on training data but poorly on test or validation data because it fails to generalize.
* **Example:** A deep decision tree that classifies every training point perfectly but fails on new inputs is overfitted.

**Detection:**

* High accuracy on training data but low accuracy on validation/test data.
* Large gap between training and validation error.

**Prevention Techniques:**

1. **Cross-Validation:** Use techniques like *k-fold cross-validation* to test model performance on multiple subsets of data.
2. **Regularization:** Apply methods like *L1 (Lasso)* and *L2 (Ridge)* regularization to penalize overly complex models.
3. **Pruning:** In decision trees, prune unnecessary branches that add complexity.
4. **Dropout:** In neural networks, randomly drop neurons during training to prevent dependency on specific features.
5. **Early Stopping:** Stop training once validation error starts increasing, even if training error continues to decrease.

---

**2. Underfitting**

* **Definition:** Underfitting happens when a model is **too simple** to capture the underlying patterns of the data. It fails to learn sufficiently during training.
* **Effect:** The model performs poorly on both training and test data because it cannot represent the complexity of the problem.
* **Example:** A linear regression model used for a non-linear dataset may underfit the data.

**Detection:**

* High error on both training and validation sets.
* Model predictions show a lack of sensitivity to changes in input data.

**Prevention Techniques:**

1. **Increase Model Complexity:** Use more advanced algorithms or add more features.
2. **Reduce Regularization:** Excessive regularization can overly constrain the model.
3. **Feature Engineering:** Add or transform input features to better represent the underlying patterns.
4. **Train Longer:** Ensure the model is trained adequately to minimize bias.

---

 **3. Bias-Variance Tradeoff**

* The **bias-variance tradeoff** explains the balance between **underfitting** and **overfitting**:

  * **High Bias (Underfitting):** The model is too simple and cannot capture complexity.
  * **High Variance (Overfitting):** The model is too complex and sensitive to small changes in data.
* The goal is to find the **optimal point** where both bias and variance are minimized, resulting in the best generalization performance.

---

 **4. Summary Table**

| Concept                    | Description                                        | Detection                                   | Prevention                                                |
| -------------------------- | -------------------------------------------------- | ------------------------------------------- | --------------------------------------------------------- |
| **Overfitting**            | Model too complex; memorizes training data         | High training accuracy, low test accuracy   | Regularization, dropout, pruning, cross-validation        |
| **Underfitting**           | Model too simple; fails to learn patterns          | High error on both training & test sets     | Increase complexity, feature engineering, longer training |
| **Bias-Variance Tradeoff** | Balancing model complexity for best generalization | Analyzing training & validation performance | Optimal model selection                                   |




### **Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.**

Hint: Consider deletion, mean/median imputation, and predictive modeling.

 **Handling Missing Values in a Dataset**

Missing values are common in datasets and can lead to inaccurate analysis or biased model predictions if not treated properly. Before choosing a method, it’s important to identify the pattern and reason for missing data — whether it’s **Missing Completely at Random (MCAR)**, **Missing at Random (MAR)**, or **Not Missing at Random (NMAR)**. Once understood, we can use various strategies to handle them effectively. Three common approaches are **deletion**, **imputation**, and **predictive modeling**.

---

 **1. Deletion Method**

The simplest method is **deletion**, where we remove rows or columns that contain missing values.

* **Listwise deletion** removes the entire row if any column has a missing value.
* **Pairwise deletion** only removes missing values for the specific analysis being performed.

**Example:**
If a dataset of 1,000 customers has 5 rows with missing income data, we can delete those rows using Pandas in Python:

```python
df.dropna(inplace=True)
```

This method is efficient when the proportion of missing data is small (<5%). However, deleting too many rows can lead to loss of valuable information and introduce bias if the missing data isn’t random.

---

 **2. Mean/Median Imputation**

Another widely used method is **imputation**, where we fill in missing values with estimated data. The most common techniques are **mean**, **median**, or **mode** imputation depending on the type of variable.

* Use **mean** for normally distributed numerical data.
* Use **median** for skewed data to reduce the effect of outliers.
* Use **mode** for categorical variables.

**Example:**
If the “Age” column has some missing values:

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

This replaces missing ages with the average age. Though simple and fast, this method can reduce data variability and may not reflect true data relationships.

---

 **3. Predictive Modeling**

A more advanced method is **predictive modeling**, where missing values are estimated using machine learning models such as **Linear Regression**, **Decision Trees**, or **KNN Imputer**. The idea is to use other available features to predict the missing ones.

**Example:**
If the “Salary” column has missing values, we can use regression to predict “Salary” based on “Experience,” “Education,” and “Age.”
In Python:

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_imputed = imputer.fit_transform(df)
```

This approach preserves the underlying data relationships and improves accuracy, but it’s more computationally intensive.





### **Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).**
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.



An imbalanced dataset occurs when the number of samples in different classes is not equal. For example, in a fraud detection dataset, 95% of transactions may be non-fraudulent and only 5% fraudulent. This imbalance can cause a model to perform poorly on the minority class since it learns mostly from the majority data.



 **1. Random Sampling (Over/Under Sampling)**

This technique balances the dataset by adjusting sample sizes.

* **Oversampling** duplicates or adds more samples from the minority class.
* **Undersampling** removes some samples from the majority class.

**Example:**

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)
```

It’s simple to apply but oversampling may cause overfitting, and undersampling can remove useful data.



 **2. SMOTE (Synthetic Minority Oversampling Technique)**

**SMOTE** creates new, synthetic samples for the minority class instead of simply duplicating existing ones. It helps the model learn better decision boundaries.

Examples-

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_sm, y_sm = sm.fit_resample(X, y)
```

SMOTE is more effective than random oversampling but can sometimes create unrealistic data points.



**Conclusion**

Imbalanced datasets can bias model predictions toward the majority class. Techniques like **Random Sampling** and **SMOTE** help create a balanced dataset, improving model accuracy and fairness.




### **Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.**
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.





Feature scaling is an important preprocessing step in Machine Learning that ensures all features contribute equally to the model. Since different features can have different ranges (e.g., age in years vs. income in lakhs), models that rely on distance or gradient calculations can get biased toward features with larger values. Scaling helps improve the performance and convergence speed of algorithms like K-Nearest Neighbors (KNN)**, **Support Vector Machines (SVM), and Gradient Descent–based models
.



 **1. Min-Max Scaling**

**Min-Max Scaling** (also called normalization) transforms features to a fixed range, usually **[0, 1]**. It is given by the formula:
[
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
]
It preserves the shape of the original distribution but can be sensitive to outliers since extreme values can distort the scale.

**Example:**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

This method works well when the data is uniformly distributed and when algorithms need bounded input values, like Neural Networks or KNN.



 **2. Standardization**

**Standardization** (or Z-score scaling) rescales data so that it has a **mean of 0** and a **standard deviation of 1**. The formula is:
[
X' = \frac{X - \mu}{\sigma}
]
It is less affected by outliers and suitable for algorithms that assume normally distributed data, such as **SVM**, **Logistic Regression**, and models using **gradient descent**.

**Example:**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```



**Conclusion**

In summary, **feature scaling** ensures that all features are on the same scale, improving model training and accuracy. **Min-Max Scaling** is ideal for data within a known range, while **Standardization** is preferred when data contains outliers or follows a normal distribution. Both techniques make models more stable, faster, and accurate.



### **Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**

Hint: Consider categorical variables with ordinal vs. nominal relationships.

Answer: Both Label Encoding and One-Hot Encoding are techniques used to convert categorical data into a numerical format that can be used by machine learning algorithms. However, they differ in how they represent the categorical values and are suitable for different types of categorical data.

Label Encoding:

- Explanation: Assigns a unique integer to each category in a feature. For example, if a "color" feature has categories "Red", "Green", and "Blue", Label Encoding might convert them to 0, 1, and 2 respectively.
- When to Prefer: Best used when the categorical variable has an ordinal relationship, meaning there is a natural order or ranking among the categories (e.g., "Low", "Medium", "High"). In this case, the numerical representation reflects the order. Using Label Encoding for nominal data can mislead algorithms into assuming a false order or relationship between categories.
- Example: Education levels (High School, Bachelor's, Master's, PhD), size categories (Small, Medium, Large).

One-Hot Encoding:

- Explanation: Creates a new binary column for each unique category in the original feature. If a category is present for a given data point, the corresponding new column will have a value of 1, and all other category columns will have a value of 0. For the "color" example with "Red", "Green", and "Blue", One-Hot Encoding would create three new columns: "Color_Red", "Color_Green", and "Color_Blue". A data point with "Red" would have Color_Red=1, Color_Green=0, Color_Blue=0.
- When to Prefer: Best used when the categorical variable has a nominal relationship, meaning there is no inherent order or ranking among the categories (e.g., colors, cities, types of fruits). One-Hot Encoding avoids creating a false sense of order or numerical relationship between categories, which can happen with Label Encoding on nominal data.
- Example: Cities (New York, London, Paris), product categories (Electronics, Clothing, Books).

### **Question 9: Flight Price Prediction Dataset**
**a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.**

**b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?**

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

a) Price vs. Days Left:

- Flight prices generally decrease when booked earlier and increase exponentially as departure nears (especially within 5–10 days).

- Best time to book: 20–30 days before departure.

b) Airline Comparison (Delhi–Mumbai):

- Cheapest: Typically low-cost airlines such as IndiGo or SpiceJet.

- Premium: Airlines like Vistara or Air India, due to added services (meals, legroom, baggage).

````python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://github.com/MasteriNeuron/datasets/raw/main/flight_price.csv"
df = pd.read_csv(url)

# Display first few rows
print("Dataset Preview:")
print(df.head())

# Basic info
print("\nDataset Info:")
print(df.info())

# Clean and preprocess data
df = df.dropna(subset=['price', 'days_left'])  # Remove rows with missing price or days_left

# ---------------------------
# (a) PRICE VARIATION WITH DAYS LEFT
# ---------------------------

# Group by days_left and calculate average price
price_trend = df.groupby('days_left')['price'].mean().reset_index()

# Plot price vs days_left
plt.figure(figsize=(10,6))
sns.lineplot(x='days_left', y='price', data=price_trend, marker='o')
plt.title('Average Flight Price vs Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (INR)')
plt.grid(True)
plt.show()

# Identify exponential surge pattern
# We'll check the correlation between log(days_left) and price
price_trend['log_days'] = np.log(price_trend['days_left'] + 1)
correlation = price_trend['log_days'].corr(price_trend['price'])
print(f"\nCorrelation between log(days_left) and price: {correlation:.2f}")

# Recommendation for best booking window
best_window = price_trend.loc[price_trend['price'].idxmin(), 'days_left']
print(f"\n✅ Recommended best booking window: Around {best_window} days before departure for lowest prices.")

# ---------------------------
# (b) PRICE COMPARISON ACROSS AIRLINES (Delhi–Mumbai)
# ---------------------------

# Filter Delhi–Mumbai route
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Average price by airline for this route
airline_prices = route_df.groupby('airline')['price'].mean().reset_index().sort_values('price')

print("\nAverage Prices for Delhi–Mumbai Route:")
print(airline_prices)

# Plot airline price comparison
plt.figure(figsize=(10,6))
sns.barplot(x='price', y='airline', data=airline_prices, palette='viridis')
plt.title('Average Flight Prices by Airline (Delhi–Mumbai)')
plt.xlabel('Average Price (INR)')
plt.ylabel('Airline')
plt.show()

# Identify cheapest and premium airlines
cheapest = airline_prices.iloc[0]
premium = airline_prices.iloc[-1]
print(f"\n✈️ Cheapest airline: {cheapest['airline']} (Avg ₹{cheapest['price']:.0f})")
print(f"💼 Most premium airline: {premium['airline']} (Avg ₹{premium['price']:.0f})")

# Reasoning
print("""
Analysis:
- Prices tend to rise sharply as the departure date gets closer (especially within the last 5–10 days).
- The exponential surge occurs because airlines increase prices for last-minute travelers.
- Booking 20–30 days in advance generally gives the best deals.
- Low-cost carriers like SpiceJet or IndiGo are usually cheaper,
  while premium airlines like Vistara or Air India offer higher prices due to better service, food, and luggage benefits.
""")
````
